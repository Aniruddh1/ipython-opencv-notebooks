#### Image Processing Analysis
- Stats (mean, stdev, variance) of Velocity, Vol_top and Vol_bot of each column

```python
col_lst = ['Top Mean', 'Bot Mean']
for col_name in col_lst:
    x = image_stats_df[[col_name]].values.astype(float)
    min_max_scaler = preprocessing.MinMaxScaler()
    image_stats_df[col_name+' Scaled'] = pd.DataFrame(min_max_scaler.fit_transform(x))
image_stats_df.head()
```

```python
# looks good in dark mode:
plt.style.use(['seaborn-dark-palette'])
fig, axes = plt.subplots(2, 1, figsize=(30, 20))

for col_name in ['Top Mean', 'Bot Mean']:
    image_stats_df[col_name].plot(ax=axes[0], y=col_name, label=col_name)
axes[0].set_ylabel("Image Intensity - Mean Pixel Value (0-65K)")
axes[0].set_title("Image Intensity - Mean Pixel Value (0-65K)")
axes[0].legend()

for col_name in ['Top Saturation', 'Bot Saturation']:
    image_stats_df[col_name].plot(ax=axes[1], y=col_name, label=col_name)
axes[1].set_ylabel("Image Saturation - Percent")
axes[1].set_title("Image Saturation - Percent")
axes[1].legend()

plt.savefig(f"{result_path}/image_stats_plots.png")
plt.show()
```
