```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

import h5py

import sys
sys.path.append('../scripts')
import util

import os
cwd=os.getcwd()
print("cwd = " + cwd)
```

### Read in jpeg of Galaxy M51

```python
m51 = cv2.imread(cwd+'/../images/m51.jpg',cv2.IMREAD_COLOR)
m51.dtype
```

### Create HPF5 file

```python
m51_h5 = h5py.File("m51.hdf5", "w")
```

### Create a data set in the HPF5 file
#### Note the shape and dtype args are set to the same as the m51 image

```python
dset = m51_h5.create_dataset("m51_ds2", m51.shape, m51.dtype)
```

```python
dset.shape
dset.dtype
dset.name
```

### Write an attribute...

```python
dset = m51_h5.get("m51_ds2")
dset.attrs['galaxy'] = 'M51'
```

### Write data to the dataset using slicing.
#### m51 is a numpy array.  'dset[...]' is the dataset data...

```python
dset[...] = m51
```

```python
m51_h5.close()
```
