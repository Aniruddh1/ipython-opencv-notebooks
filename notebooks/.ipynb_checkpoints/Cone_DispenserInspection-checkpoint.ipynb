```python
sample_image_path = "../images/Dispenser Defects/"
```

```python
img_path_list = [str(img_path) for img_path in Path(sample_image_path).glob('**/*.PNG')]
print(f"Found {len(img_path_list)} images...")

categories = ['artifact', 'channel', 'tooling', 'debris', 'bubble']

img_categories = {}
img_categories['other'] = []

for img_path in img_path_list:
    categorized = False
    for category in categories:
        if os.path.basename(img_path).lower().find(category) >= 0:
            if category not in img_categories:
                img_categories[category] = []
            img_categories[category].append(img_path)
            categorized = True
            break
    if not categorized:
        img_categories['other'].append(img_path)
```

```python
for category in img_categories.keys():
    print(f"Images from category '{category}'")
    imagesList=''.join( ["<img style='width: 120px; margin: 1px; float: left; border: 2px solid black;' src='%s' />" % str(s) 
                 for s in sorted(img_categories[category]) ])
    display(HTML(imagesList))
```

```python
fname = img_categories['other'][-2]
img = cv2.imread(fname, cv2.IMREAD_ANYCOLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_gray.shape
plt.imshow(img_gray, cmap=plt.cm.gray, interpolation='nearest')
plt.title(f'Image path: {fname}')
plt.axis('off')
plt.show()
```

```python
p_x = img_gray.mean(axis=0)
p_x_grad = np.gradient(p_x)
p_x_grad = p_x_grad-np.min(p_x_grad)
#plt.plot(p_x)
plt.plot(p_x_grad)
max_idx = np.argmax(p_x_grad)
min_idx = np.argmin(p_x_grad)
print("X: min idx: %d, max idx: %d" % (min_idx, max_idx))
x = min_idx + int(0.5 * (max_idx - min_idx))

mean = np.mean(p_x_grad)
print(f"Mean: {mean}")

p_x_bin = p_x_grad > (mean+(mean*0.5))
plt.figure()
plt.plot(p_x_bin)
```
