```python
def zminmaxscore(r, key):
    try:
        min = r.zrangebyscore(key, '-inf', '+inf', withscores=True, start=1, num=1)[0][1]
        max = r.zrevrangebyscore(key, '+inf', '-inf', withscores=True, start=1, num=1)[0][1]
        return (min, max)
    except Exception as e:
        print( str(e) )
        return tuple()

# utils specific to DIF
#
def parseExpRoot(hash_in):
    # expRoot: "ExperimentID":run_num:"HeadSerialNumber":"NozzleSize":"MonomerID"
    return ':'.join(hash_in.split(':')[0:5])

def filter_list_of_res_by_nozzle(result_list, nozzle_of_interest):
    if nozzle_of_interest == -1:
        return result_list
    
    nozzle_filter = "res:1:%d:" % nozzle_of_interest
    filtered_results = []
    for res in result_list:
        if res[0].find(nozzle_filter) > -1:
            filtered_results.append(res)
    return filtered_results

def getNozzlePhase(nozzle_num):
    return ["B", "C", "A"][(nozzle_num%3)]
```

```python
completed_list = list(r.smembers('experiments_completed'))
exp_root = completed_list[0]
vel_hash = "%s:res_vel_set" % (exp_root)
volume_hash = "%s:res_vol_set" % (exp_root)

print (vel_hash)
print (volume_hash)
```

```python
#  Num of bins to separate the scores into
bin_cnt = 30

#  Set this to nozzle num to filter out all other nozzles, or
#  to -1 to turn filtering off
nozzle_of_interest = 250
```

```python
#  Find the min, max scores for volume and velocity
#
vol_min_max = zminmaxscore(r, volume_hash)
vel_min_max = zminmaxscore(r, vel_hash)

#  From the min, max values, calc bin sizes and range of scores
vol_score_range = vol_min_max[1] - vol_min_max[0]
vol_bin_size = vol_score_range /  bin_cnt
vel_score_range = vel_min_max[1] - vel_min_max[0]
vel_bin_size  = vel_score_range /  bin_cnt

print("Velocity min: %f, max: %f" % (vel_min_max[0], vel_min_max[1]))
print("Volume min  : %f, max: %f" % (vol_min_max[0], vol_min_max[1]))
```

```python
#  Using bin size and ranges, separate velocity and volume data into
#  bins (contained by vol_bin_sets).
#
vol_bin_sets = []                       # list of lists of zranges (tuple of res_hash, score)
vol_bin_ranges = []                     # list of tuples of low, high of bin range 
vol_hist_bins = []                      # list of threshold values for the bins
vol_bin_set_counts = []                 # list of lengths (count) of the bins
vol_cnt = 0                             # overall count of all results
print("Volume bins:")
for idx in range(bin_cnt):
    low = vol_min_max[0]+(idx*vol_bin_size)
    high = low + vol_bin_size
    zrange_lst = r.zrangebyscore(volume_hash, low, high, withscores=True)
    
    zrange_lst_filtered = filter_list_of_res_by_nozzle(zrange_lst, nozzle_of_interest)
    
    vol_cnt += len(zrange_lst_filtered)
    vol_hist_bins.append( low+((high-low)/2.0)  )
    vol_bin_ranges.append( (low, high) )
    vol_bin_sets.append( zrange_lst_filtered )
    vol_bin_set_counts.append( len(zrange_lst_filtered) )
    print(" low = %f, high = %f, count=%d" % (low, high, len(zrange_lst_filtered)) )
print("Volume total: %d\n-----------------------------------------" % (vol_cnt))
#vol_hist_bins.append(high)

vel_bin_sets = []
vel_bin_ranges = []
vel_hist_bins = []
vel_bin_set_counts = []
vel_cnt = 0
print("Velocity bins:")
for idx in range(bin_cnt):
    low = vel_min_max[0]+(idx*vel_bin_size)
    high = low + vel_bin_size
    zrange_lst = r.zrangebyscore(vel_hash, low, high, withscores=True)
    
    zrange_lst_filtered = filter_list_of_res_by_nozzle(zrange_lst, nozzle_of_interest)
    
    vel_cnt += len(zrange_lst_filtered)
    vel_hist_bins.append(low+((high-low)/2.0))
    vel_bin_ranges.append( (low, high) )
    vel_bin_sets.append( zrange_lst_filtered )
    vel_bin_set_counts.append( len(zrange_lst_filtered) )
    print(" low = %f, high = %f, count=%d" % (low, high, len(zrange_lst_filtered)) )
print("Velocity total: %d\n-----------------------------------------" % (vel_cnt))
#vel_hist_bins.append(high)
```

```python
#
#  Get all the

out_file = "%s/%s_vel_variance.txt" % (paths.tmp_dir, exp_root)
print(exp_root)
out_file = out_file.replace(":", "-")
f = open(out_file, 'w')

vel_variance_by_loop_zset = "%s:res_vel_variance" % (exp_root)
vol_variance_by_loop_zset = "%s:res_vol_variance" % (exp_root)

r.delete(vel_variance_by_loop_zset)
r.delete(vol_variance_by_loop_zset)

vel_variance_lst = []
# velA_variance_lst = []
# velB_variance_lst = []
# velC_variance_lst = []
vol_variance_lst = []
# volA_variance_lst = []
# volB_variance_lst = []
# volC_variance_lst = []

n2n_vel_vars_global = {"A-to-C":[], "A-to-B":[], "B-to-C":[]}
n2n_vol_vars_global = {"A-to-C":[], "A-to-B":[], "B-to-C":[]}

desired_row=1
for vel_desired in vel_range_desired:
    h = vel_desired[0]
    loop_num = h[h.rfind(':')+1:]
    exp_root = parseExpRoot(h)
    base = '%s:res:%d:*' % (exp_root, desired_row)
    vel_desired_loop_num_set_unsorted = r.keys(pattern='%s:%s' % (base, loop_num))
    vel_desired_loop_num_set_sorted = sorted(vel_desired_loop_num_set_unsorted)
    
    loop_vars = r.hget("%s:vars:%s" % (exp_root, loop_num), "Nozzle")
    NOI = int(float(r.hget("%s:vars:%s" % (exp_root, loop_num), "Nozzle")))
    phase_of_NOI = getNozzlePhase(NOI)
    
    vels = {"All":[], "A":[], "B":[], "C":[]}
    vols = {"All":[], "A":[], "B":[], "C":[]}
    
    pop_size = len(vel_desired_loop_num_set_sorted)
    for idx, res_hash in enumerate(vel_desired_loop_num_set_sorted):
        res = r.hgetall(res_hash)
        vels["All"].append(float(res['vel']))
        vols["All"].append(float(res['volume']))
        phase = getNozzlePhase(int(res['Nozzle']))
        
        vels[phase].append(float(res['vel']))
        vols[phase].append(float(res['volume']))
        
        NOI_flag = "*" if nozzle_of_interest == int(res['Nozzle']) else " "
        f.write("%s%s:%s: Velocity = %.3f, Volume = %.3f\n" % 
              (NOI_flag, res_hash, phase, float(res['vel']), float(res['volume'])) )
    
    # insert a nan if lst is empty so math below doesn't throw exception
    if len(vels["A"]) == 0: vels["A"].append(np.nan)
    if len(vels["B"]) == 0: vels["B"].append(np.nan)
    if len(vels["C"]) == 0: vels["C"].append(np.nan)
    if len(vols["A"]) == 0: vels["A"].append(np.nan)
    if len(vols["B"]) == 0: vels["B"].append(np.nan)
    if len(vols["C"]) == 0: vels["C"].append(np.nan)
        
    n2n_vel_vars = []
    n2n_vol_vars = []
    
    if False:
        if phase_of_NOI == "A":
            n2n_vel_vars.append( ("A-to-B", np.var( [np.array(vels["A"]).mean(), np.array(vels["B"]).mean()])) )
            n2n_vel_vars.append( ("A-to-C", np.var( [np.array(vels["A"]).mean(), np.array(vels["C"]).mean()])) )
            n2n_vol_vars.append( ("A-to-B", np.var( [np.array(vols["A"]).mean(), np.array(vols["B"]).mean()])) )
            n2n_vol_vars.append( ("A-to-C", np.var( [np.array(vols["A"]).mean(), np.array(vols["C"]).mean()])) )
        elif phase_of_NOI == "B":
            n2n_vel_vars.append( ("B-to-A", np.var( [np.array(vels["B"]).mean(), np.array(vels["A"]).mean()])) )
            n2n_vel_vars.append( ("B-to-C", np.var( [np.array(vels["B"]).mean(), np.array(vels["C"]).mean()])) )
            n2n_vol_vars.append( ("B-to-A", np.var( [np.array(vols["B"]).mean(), np.array(vols["A"]).mean()])) )
            n2n_vol_vars.append( ("B-to-C", np.var( [np.array(vols["B"]).mean(), np.array(vols["C"]).mean()])) )
        elif phase_of_NOI == "C":
            n2n_vel_vars.append( ("C-to-A", np.var( [np.array(vels["C"]).mean(), np.array(vels["A"]).mean()])) )
            n2n_vel_vars.append( ("C-to-B", np.var( [np.array(vels["C"]).mean(), np.array(vels["B"]).mean()])) )
            n2n_vol_vars.append( ("C-to-A", np.var( [np.array(vols["C"]).mean(), np.array(vols["A"]).mean()])) )
            n2n_vol_vars.append( ("C-to-B", np.var( [np.array(vols["C"]).mean(), np.array(vols["B"]).mean()])) )
    else:
        # velocities
        if not np.isnan(np.array(vels["A"]).mean()) and not np.isnan(np.array(vels["B"]).mean()):
            v = np.var( [np.array(vels["A"]).mean(), np.array(vels["B"]).mean()])
            n2n_vel_vars.append( ("A-to-B", v) )
            n2n_vel_vars_global["A-to-B"].append(v)
                                
        if not np.isnan(np.array(vels["A"]).mean()) and not np.isnan(np.array(vels["C"]).mean()):
            v = np.var( [np.array(vels["A"]).mean(), np.array(vels["C"]).mean()])
            n2n_vel_vars.append( ("A-to-C", v) )
            n2n_vel_vars_global["A-to-C"].append(v)
   
        if not np.isnan(np.array(vels["B"]).mean()) and not np.isnan(np.array(vels["C"]).mean()):
            v = np.var( [np.array(vels["B"]).mean(), np.array(vels["C"]).mean()])
            n2n_vel_vars.append( ("B-to-C", v) )
            n2n_vel_vars_global["B-to-C"].append(v)

        # volumes
        if not np.isnan(np.array(vols["A"]).mean()) and not np.isnan(np.array(vols["B"]).mean()):
            v = np.var( [np.array(vols["A"]).mean(), np.array(vols["B"]).mean()])
            n2n_vol_vars.append( ("A-to-B", v) )
            n2n_vol_vars_global["A-to-B"].append(v)
                                
        if not np.isnan(np.array(vols["A"]).mean()) and not np.isnan(np.array(vols["C"]).mean()):
            v = np.var( [np.array(vols["A"]).mean(), np.array(vols["C"]).mean()])
            n2n_vol_vars.append( ("A-to-C", v) )
            n2n_vol_vars_global["A-to-C"].append(v)
   
        if not np.isnan(np.array(vols["B"]).mean()) and not np.isnan(np.array(vols["C"]).mean()):
            v = np.var( [np.array(vols["B"]).mean(), np.array(vols["C"]).mean()])
            n2n_vol_vars.append( ("B-to-C", v) )
            n2n_vol_vars_global["B-to-C"].append(v)

    vol_var = np.var(vols["All"])
    vel_var = np.var(vels["All"])
    f.write("Velocity Variance:\n")
    f.write("  All: %.3f, Just A: %.3f, Just B: %.3f, Just C: %.3f\n" % (vel_var, np.var(vels["A"]), np.var(vels["B"]), np.var(vels["C"])) )
    f.write("  nozzle-to-nozzle (phase of NOI: %s):\n" % (phase_of_NOI))
    for n2n_vel_var in n2n_vel_vars:
        f.write("   %s: %.3f\n" % (n2n_vel_var[0], n2n_vel_var[1]) )
    f.write("Volume Variance:\n")
    f.write("  All: %.3f, Just A: %.3f, Just B: %.3f, Just C: %.3f\n" % (vol_var, np.var(vols["A"]), np.var(vols["B"]), np.var(vols["C"])) )
    f.write("  nozzle-to-nozzle (phase of NOI: %s):\n" % (phase_of_NOI))
    for n2n_vol_var in n2n_vol_vars:
        f.write("   %s: %.3f\n" % (n2n_vol_var[0], n2n_vol_var[1]) )
    f.write("------------------------------------------------------------------------------------------------\n")

    r.zadd(vel_variance_by_loop_zset, vel_var, "%s:%d" % (loop_num, pop_size))
    r.zadd(vol_variance_by_loop_zset, vol_var, "%s:%d" % (loop_num, pop_size))
    
    vel_variance_lst.append(vel_var)
    vol_variance_lst.append(vol_var)

f.close()
```

```python
def reject_outliers_no(data, m = 2.):
    #https://stackoverflow.com/questions/11686720/is-there-a-numpy-builtin-to-reject-outliers-from-a-list
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]

def filter_outliers(data):
    mean = np.mean(data, axis=0)
    sd = np.std(data, axis=0)
    inter = [x for x in data if (x > mean - 2 * sd)]
    return [x for x in inter if (x < mean + 2 * sd)]

vel_variance_np = np.array(vel_variance_lst)
vol_variance_np = np.array(vol_variance_lst)

vel_variance_filtered_np = filter_outliers(vel_variance_np)
vol_variance_filtered_np = filter_outliers(vol_variance_np)

fig = plt.figure(figsize=(30,20))
plt.clf()
ax1 = fig.add_subplot(211)
#ax2 = ax1.twinx()
ax2 = fig.add_subplot(212)

ax1.plot(vel_variance_filtered_np,  color='b', label='Velocity')
ax2.plot(vol_variance_filtered_np,  color='r', label='Volume')

ax1.set_xlabel('Inter Score')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Velocity', color='b')
ax1.tick_params('y', colors='b')
ax2.set_ylabel('Volume', color='r')
ax2.tick_params('y', colors='r')

leg1 = ax1.legend(loc='upper left', shadow=True)
leg2 = ax2.legend(loc='upper right', shadow=True)

#plt.ylim([y_loc,22])
plt.title("Variance: Volume and Velocity")
#plt.tight_layout()
plt.show()

print("Vel: filtered %d outliers" % (len(vel_variance_np) - len(vel_variance_filtered_np)))
print("Vol: filtered %d outliers" % (len(vol_variance_np) - len(vol_variance_filtered_np)))
# for i in range(len(vel_variance_lst)):
#     if vel_variance_np[i] != vel_variance_filtered_np[i]:
#         print("filtered idx: %d (orig: %.3f, filtered: %)" % (i))
```
