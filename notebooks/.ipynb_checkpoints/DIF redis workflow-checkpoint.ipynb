#### Determine distribution of volume and velocity

Create histograms for volume and velocity data

```python
#  Num of bins to separate the scores into
bin_cnt = 30

#  Set this to row-nozzle pair to filter out all other nozzles, or
#  to '*:*' to turn filtering off
#  row_noz_pair is a string representation of a float where the integer portion is the row and the decimal portion is the nozzle
#  ie: 
#    '1.455' == nozzle 455 of row 1 (Xaar 100x)
#    '2.2' == nozzle 2 of row 2 (Xaar 100x or OCE)
#    '3.82' == nozzle 82 of row 3 (OCE)

row_noz_pair = '*'
#row_noz_pair = '1:250'
```

```python
#  Find the min, max scores for volume and velocity
#
vol_min_max = du.zminmaxscore(r, res_by_vol_zset_key)
vel_min_max = du.zminmaxscore(r, res_by_vel_zset_key)

#  From the min, max values, calc bin sizes and range of scores
vol_score_range = vol_min_max[1] - vol_min_max[0]
vol_bin_size = vol_score_range /  bin_cnt
vel_score_range = vel_min_max[1] - vel_min_max[0]
vel_bin_size  = vel_score_range /  bin_cnt

print("Velocity min: %f, max: %f" % (vel_min_max[0], vel_min_max[1]))
print("Volume min  : %f, max: %f" % (vol_min_max[0], vol_min_max[1]))
```

```python
#  Using bin size and ranges, separate velocity and volume data into bins:
#
vol_bin_sets = []                       # list of lists of zranges (tuple of res_key, score)
vol_bin_ranges = []                     # list of tuples of low, high of bin range 
vol_hist_bins = []                      # list of threshold values for the bins
vol_bin_set_counts = []                 # list of lengths (count) of the bins
vol_cnt = 0                             # overall count of all results
print("Volume bins:")
for idx in range(bin_cnt):
    low = vol_min_max[0]+(idx*vol_bin_size)
    high = low + vol_bin_size
    zrange_lst = r.zrangebyscore(res_by_vol_zset_key, low, high, withscores=True)
    
    zrange_lst_filtered = filter_list_of_res_by_nozzle(zrange_lst, row_noz_pair)
    
    vol_cnt += len(zrange_lst_filtered)
    vol_hist_bins.append( low+((high-low)/2.0)  )
    vol_bin_ranges.append( (low, high) )
    vol_bin_sets.append( zrange_lst_filtered )
    vol_bin_set_counts.append( len(zrange_lst_filtered) )
    print("[bin %2d] low = %f, high = %f, count=%d" % (idx+1, low, high, len(zrange_lst_filtered)) )
print("Volume total: %d\n-----------------------------------------" % (vol_cnt))
#vol_hist_bins.append(high)

vel_bin_sets = []
vel_bin_ranges = []
vel_hist_bins = []
vel_bin_set_counts = []
vel_cnt = 0
print("Velocity bins:")
for idx in range(bin_cnt):
    low = vel_min_max[0]+(idx*vel_bin_size)
    high = low + vel_bin_size
    zrange_lst = r.zrangebyscore(res_by_vel_zset_key, low, high, withscores=True)
    
    zrange_lst_filtered = filter_list_of_res_by_nozzle(zrange_lst, row_noz_pair)
    
    vel_cnt += len(zrange_lst_filtered)
    vel_hist_bins.append(low+((high-low)/2.0))
    vel_bin_ranges.append( (low, high) )
    vel_bin_sets.append( zrange_lst_filtered )
    vel_bin_set_counts.append( len(zrange_lst_filtered) )
    print("[bin %2d] low = %f, high = %f, count=%d" % (idx+1, low, high, len(zrange_lst_filtered)) )
print("Velocity total: %d\n-----------------------------------------" % (vel_cnt))
#vel_hist_bins.append(high)
```

### Rename result keys / refactor

Recently added a new hash to capture the dispense parameters used in a loop: DPID_x, where x is 
a sequential number.  
For (future) result processing, it will be easier to have the new dispense parameter as part of
the result key.  The cells below are for renaming the existing keys to add the DPID to the key.

```python
'''
Get all result keys

First do a full scan to get all result keys: exp_root:res:*
'''
result_keys_all = []
cur, keys = r.scan(match='%s:res:*' % exp_root)
result_keys_all += keys
#print(cur, len(keys))
while cur > 0:
    #for i in range(5):
    cur, keys = r.scan(cursor=cur, match='%s:res:*' % exp_root)
    #print(cur, len(keys))
    result_keys_all += keys
len(result_keys_all)

# Ooops.  because of a bug in the above snippet, there were a handfull of results which did not get included in result_keys_all, and therefore did not get the DPID added to the key.  
# The snippet has been fixed, but the code below was used to identify result keys which did not have DPID added.  The cell below is a special version of the cell used to rename the 
# result keys and is used only to process the result keys in 'missing' list (which should now be empty)
missing = []
for result_key in result_keys_all:
    if result_key.find('DPID') < 0:
        missing.append(result_key)
# if this is empty, then we are good...
print(missing)
```

```python
'''
HACK ALERT!! This is a copy of the original cell below, but only for processing the list of 'missing' result keys...see above.

Rename all result keys with new dispenser parameter ID embedded in key, and add new hash 
for each unique dispenser parameter set.
'''
disp_param_set_key = f"{exp_root}:disp_param_set"

set_of_disp_params = set()
disp_params_by_hash = {}
disp_param_cnt = r.scard(disp_param_set_key)  # start where we left off...

# for this version, we need to first rebuild our local set_of_disp_params and disp_params_by_hash
existing_disp_param_set = r.smembers(disp_param_set_key)
for disp_param_id in existing_disp_param_set:
    disp_param_hash_key = f"{exp_root}:{disp_param_id}"
    disp_vars = r.hgetall(disp_param_hash_key)
    disp_parms = f"{disp_vars['BitmapFile']}:{disp_vars['Waveform']}:{disp_vars['Sample Clock']}:{disp_vars['Global Voltage']}"
    # add disp_parms to local set
    set_of_disp_params.add(disp_parms)
    disp_params_by_hash[disp_parms] = disp_param_id

for result_key in missing:
    # get loop count from key and add to result hash
    loop_count = result_key[result_key.rfind(':')+1:]
    r.hset(result_key, 'Loop_Count', loop_count)
    
    # create string-hash from the values in results that pertain to dispense params
    disp_vars = r.hgetall(result_key)
    disp_parms = f"{disp_vars['BitmapFile']}:{disp_vars['Waveform']}:{disp_vars['Sample Clock']}:{disp_vars['Global Voltage']}"

    # See if we've come across these disp params before? 
    # if so, get disp_param_id from dict
    # if not:
    #  - create unique DPID using count and add to dict 
    #  - add to set
    #  - add to redis set of all DPID's
    #  - create a redis hash with the disp params
    if disp_parms not in set_of_disp_params:
        print(f"Creating new DPID for parms: {disp_parms}")
        
        # create unique DPID using count
        disp_param_cnt += 1
        disp_param_id = f"DPID_{disp_param_cnt}"
        disp_params_by_hash[disp_parms] = disp_param_id
        
        # add disp_parms to local set
        set_of_disp_params.add(disp_parms)
        
        # add to redis set of all DPID's
        r.sadd(disp_param_set_key, disp_param_id)
        
        # create new hash using DPID in key with all the disp settings
        disp_param_hash_key = f"{exp_root}:{disp_param_id}"
        r.hset(disp_param_hash_key, 'BitmapFile', disp_vars['BitmapFile'])
        r.hset(disp_param_hash_key, 'Waveform', disp_vars['Waveform'])
        r.hset(disp_param_hash_key, 'Sample Clock', disp_vars['Sample Clock'])
        r.hset(disp_param_hash_key, 'Global Voltage', disp_vars['Global Voltage'])
    else:
        disp_param_id = disp_params_by_hash[disp_parms]
    
    result_key_base = result_key[:result_key.rfind(':')]
    print("renamed to:", f"{result_key_base}:{disp_param_id}:{loop_count}")
    r.rename(result_key, f"{result_key_base}:{disp_param_id}:{loop_count}")
        
```

```python
'''
Okay, now we need to rebuild the sets/lists/zset/etc that reference the old result keys.

    1) Fetch contents of each set/zset/list that references any of the old style keys
    2) Iterate through each set/zset/list and build new set/zset/list (append '_NEW') using the new result key
    3) Confirm and then finally rename new set/zset/list to orig names
'''

def getNewResultKey(result_key):
    k = f"{':'.join(result_key.split(':')[0:8])}:DPID_*:{result_key.split(':')[-1]}"
    new_keys = r.keys(k)
    if len(new_keys) > 1:
        raise Exception(f"More than one result found for {k}")
    return new_keys[0]

# keys for any set/zset/list that references old result keys
all_results_set_key = "results_set"
res_by_vel_zset_key = f"{exp_root}:res_by_vel_zset"
res_by_vol_zset_key = f"{exp_root}:res_by_vol_zset"
desired_res_by_vel_zset_key = f"{exp_root}:desired_res_by_vel_zset"
desired_res_by_vol_zset_key = f"{exp_root}:desired_res_by_vol_zset"

all_results_list = list(r.smembers(all_results_set_key))
print(f"all_results_list len: {len(all_results_list)}")

# # first deal with the list
# print("------------------------------------------------")
# for result_key in all_results_list:
#     new_key = getNewResultKey(result_key)
#     #print(f"{result_key} -> {new_key}")
#     r.sadd(f"{all_results_set_key}_NEW", getNewResultKey(result_key))

# Next, deal with the zsets...
zsets_list = [res_by_vel_zset_key, res_by_vol_zset_key, desired_res_by_vel_zset_key, desired_res_by_vol_zset_key ]
zsets_list = [ desired_res_by_vel_zset_key, desired_res_by_vol_zset_key ]

for zset_key in zsets_list:
    zset = r.zrange(zset_key, 0, -1, withscores=True)
    print("------------------------------------------------")
    print(f"{zset_key} len: {len(zset)}")
    
    for result_key, score in zset:
        try:
            new_key = getNewResultKey(result_key)
        except IndexError:
            print(f"Error processing {result_key}")
            
        #print(f"{result_key} -> {new_key}")
        r.zadd(f"{zset_key}_NEW", score, new_key)
```
