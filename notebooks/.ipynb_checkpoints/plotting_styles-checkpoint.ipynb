```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris['data'], columns = list(iris['feature_names']))
# iris_df = pd.DataFrame(iris)
iris_df.head()
```

```python
# looks good in dark mode:
plt.style.use(['seaborn-dark-palette'])

iris_df.plot()
```

```python
light_plot_cnt = len(util.plt_pallets_light)
dark_plot_cnt = len(util.plt_pallets_dark)
```

```python
cols = 3
rows = int(dark_plot_cnt/cols )
fig, axes = plt.subplots(int(rows), cols, figsize=(20,10))

for r in range(rows):
    for c in range(cols):
        style = util.plt_pallets_dark[ (r*cols)+c ]
        with plt.style.context((style)):
            iris_df.plot(ax=axes[r,c])
```
