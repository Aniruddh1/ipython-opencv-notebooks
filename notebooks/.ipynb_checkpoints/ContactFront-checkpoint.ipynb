```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import cv2
import numpy as np
from matplotlib import pyplot as plt
import util
```

### Create an Image Engine to hold all the images

```python
uat = cv2.imread('/data/images/uat_24x24.png',cv2.IMREAD_GRAYSCALE)
uat_h, uat_w = uat.shape[:2]

overlay = np.full((512, 512), 255, np.uint8)

uat_locs = [(140,77), (452,464)]
uat_lst = []
for idx, uat_loc in enumerate(uat_locs):
    overlay[uat_loc[1]:uat_loc[1]+uat_h, uat_loc[0]:uat_loc[0]+uat_w] = uat
    uat_lst.append(uat)
    uat = np.rot90(uat)

image_folder_lst = ['../projects/contact_front/EFF_subset']
eng = util.Image_Engine(image_folder_lst, 'ContactFront-*.png', doSort=True, overlay=overlay)
print "Read %d images from %s" % (eng.cnt(), image_folder_lst[0])

util.plot_imgs([(eng.images[0], '0'), (eng.images[-2], '10'), (eng.images[-1], '20')])
```

```python
plot_lst = []
for img_idx, img in enumerate(eng.images):
    pts =[]
    for uat in uat_lst:
        res = cv2.matchTemplate(img, uat, cv2.TM_CCOEFF_NORMED)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
        pt = (max_loc[0] + (uat_h/2), max_loc[1] + (uat_w/2))
        pts.append(pt)
    img_roi = img[pts[0][1]:pts[1][1], pts[0][0]:pts[1][0]] #np slice: [startY:endY, startX:endX]
 
    thresh_mean = cv2.adaptiveThreshold(img_roi, 150, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 7, 3)
    thresh_gaus = cv2.adaptiveThreshold(img_roi, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 4)

    plot_lst.append( (thresh_mean, 'thresh_mean') )

    if False:
        canny = cv2.Canny(thresh_mean, 70, 120, apertureSize=3, L2gradient=False)
        _, contours, hierarchy = cv2.findContours(canny.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]

        top_peri = 0
        # loop over our contours
        for idx, c in enumerate(contours):
            peri = cv2.arcLength(c, True)
            approx = cv2.approxPolyDP(c, 0.002 * peri, True)
            if peri > top_peri:
                top_peri = peri
                top_contour = approx
    #     h, w = img.shape[:2]
    #     vis = np.zeros((h, w, 3), np.uint8)

        center, rad = cv2.minEnclosingCircle(top_contour)
        center = int(center[0]), int(center[1])
        rad = int(rad)
        cv2.circle(img, center, rad, util.green)
        plot_lst.append( (img_roi, '%d' % (img_idx)) )
        
util.plot_imgs(plot_lst)
```
