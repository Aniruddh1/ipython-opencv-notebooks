## Dispense Verification
Process flow and simple algorithm to detect if a dispense occured using before and after images.  

```python
#
# Ignore this cell, it is for notebook setup only
#
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

import sys
import paths
sys.path.append(paths.scripts_root)
import util

%load_ext autoreload
%autoreload 2

# Display entire width of browser
from IPython.display import display, HTML
display(HTML(data="""
<style>
    div#notebook-container    { width: 95%; }
    div#menubar-container     { width: 65%; }
    div#maintoolbar-container { width: 99%; }
</style>
"""))
```

### Process Flow
1. Read both images in ('before' dispense and 'after' dispense)
2. Take sub-image (ROI) to remove portion outside of mesa
3. Calculate the mean of the pixels along each row:  For each row in image, sum the pixels and divide by number of columns
4. Calculate delta between each mean vector
5. Use algorithm to determine if there was dispense or not

```python
# 1. Read both images in ('before' dispense and 'after' dispense)
before = cv2.imread(paths.images_root+'/mr_dispense/none_disp_komukai.bmp', cv2.IMREAD_GRAYSCALE)
after = cv2.imread(paths.images_root+'/mr_dispense/disp_komukai.bmp', cv2.IMREAD_GRAYSCALE)
rows, cols = before.shape
print("Rows: %d, Columns: %d" % (rows, cols))

# 2. Take sub-image (ROI) to remove portion outside of mesa
startX =31
endX   =500
startY =28
endY   =395
before_roi = before[startY:endY, startX:endX]  #np slice: [startY:endY, startX:endX]
after_roi = after[startY:endY, startX:endX]  #np slice: [startY:endY, startX:endX]

# plot images
util.plot_imgs([(before,'none_disp_komukai'), (after,'disp_komukai'),
                (before_roi,'none_disp_komukai sub-image (ROI)'), (after_roi,'disp_komukai sub-image (ROI)')], max_cols = 2)
```

```python
# 3. Calculate the mean of the pixels along each row:  For each row in image, sum the pixels and divide by number of columns
before_mean = np.mean(before_roi, axis=0)
after_mean = np.mean(after_roi, axis=0)

# 4. Calculate delta between each mean vector
delta = before_mean - after_mean
delta_mean = np.mean(delta)
print("Delta between images is %f" % (delta_mean))

# Plot vectors
plt.plot(before_mean, color='b', label='Before',)
plt.plot(after_mean, color='r', label='After')
plt.plot(delta, color='g', label='Delta')
plt.title('Before vs After Dispense')
plt.ylim([0,256])
plt.legend(loc='upper left', shadow=True)
plt.show()
```

```python
# 5. Use algorithm to determine if there was dispense or not
pixel_threshold = 15

if delta_mean > pixel_threshold:
    print("Dispense okay")
else:
    print("Error! Dispense not detected")

# POSSIBLE algorithm to detect bad/missing nozzles (dispense was okay, but with some missing nozzles)
#  This may be too simplified...
delta_min = np.min(delta)
delta_max = np.max(delta)
delta_range = delta_max - delta_min
print("Delta range: %d (Min: %d, Max: %d)" % ( delta_range, delta_min, delta_max))

spread_threshold = pixel_threshold * 0.8
if delta_range > spread_threshold:
    print("Missing nozzle(s) detected")
```
