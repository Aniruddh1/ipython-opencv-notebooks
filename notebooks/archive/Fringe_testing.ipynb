```python
i = img
# Convert BGR to HSV
hsv = cv2.cvtColor(i, cv2.COLOR_BGR2HSV)

# define range of blue color in HSV
#lower_blue = np.array([110,50,50])
#upper_blue = np.array([130,255,255])
lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])

# Threshold the HSV image to get only blue colors
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(i,i, mask= mask)

plt.subplot(1,4,1),plt.imshow(i)
plt.title('original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,2),plt.imshow(hsv)
plt.title('hsv'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,3),plt.imshow(mask)
plt.title('mask'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,4),plt.imshow(res)
plt.title('Result'), plt.xticks([]), plt.yticks([])
plt.figure(figsize=(12,4))
plt.title('Histogram for HSV image')
color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([hsv],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
