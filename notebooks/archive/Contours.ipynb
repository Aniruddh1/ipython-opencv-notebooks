```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/home/trafferty/Dropbox/work/canon_cnt/projects/fringe_detection/MRTool_fringe1.bmp',cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(1,3,1),plt.imshow(img)
plt.title('Original Color'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(img_gray,cmap = 'gray')
plt.title('Original Gray'), plt.xticks([]), plt.yticks([])
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
plt.subplot(1,3,1),plt.imshow(laplacian,cmap = 'gray')
plt.title('laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
```
