In this tutorial, we will expand on what we learned in the first image processing tutorial and develop some further skills to help us with segmentation of images of bacteria for the LacI titration experiment. The idea is to use the phase images to segment the image, identifying individual bacteria. We will then know where to look in the fluorescent images to find the fluorescent signals of each bacterium.

In the *E. coli* growth experiment, we made strong arguments for performing segmentation on images of bacteria constitutively expressing a fluorescent protein.  The main motivation is that the haloing effect from phase contrast images make segmenting bacteria in large clumps very difficult.  However, for the LacI titration experiment, we do not have a constitutive fluorescent protein.  It is a bad idea to use the fluorescent signal we are measuring to do the segementation, since this could introduce bias, especially when we have very low fluorescent signal for some cells.  So, in this experiment, it is important to have a dilute field of bacteria so that we do not have clumps of bacteria that mage segementation difficult.

Importantly, we are free to manipulate the brightfield image as we like in order to get good segmentation.  After we have identified which pixels below to which cells, we have to be very careful adjusting the fluorescent images as the pixel values in these images are the signal we are measuring.  We will only employ a median filter with a very small structuring element to deal with the known camera issue of occasional rogue high intensity pixels.

As usual, we will start by importing the modules we will need.

```python
from __future__ import division, print_function

# NumPy
import numpy as np

# SciPy ndimage
import scipy.ndimage

# Scikit-image submodules
import skimage.feature
import skimage.filters
import skimage.io
import skimage.morphology
import skimage.restoration
import skimage.segmentation

# Plotting modules
import matplotlib.pyplot as plt

# This is to enable inline displays for the purposes of the tutorial
%matplotlib inline
```

##Dealing with uneven illumination

The pair of images we will use for this tutorial are a brightfield phase contrast image of HG104 bacteria (`HG104_bf.tif`) and a fluorescent image of the same HG104 bacteria expressing YFP (`HG104_FITC.tif`).

Before we load those images, we will first learn how to deal with uneven illumination.  This can be a problem if there were sample preparation issues or if there are known problems with the optics.  We will therefore first load `HG104_bf_uneven_lighting.tif`, which has a large dark spot in the image.

```python
# Load the images
im_bf = skimage.io.imread('HG104_bf.tif')
im_fl = skimage.io.imread('HG104_FITC.tif')
im_bf_uneven = skimage.io.imread('HG104_bf_uneven_lighting.tif')

# Make sure it's an unsigned 16 bit integer (early version had type issue)
im_bf_uneven = im_bf_uneven.astype(np.uint16)

# Display the image
fig, ax = plt.subplots(1, 2, figsize=(12, 12))
ax[0].imshow(im_bf, cmap=plt.cm.gray)
ax[1].imshow(im_bf_uneven, cmap=plt.cm.gray)
ax[0].set_title('original')
ax[1].set_title('uneven illumination')
```

The effects of the uneven illumination are even more pronounced when we look with false coloring.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 12))
ax[0].imshow(im_bf, cmap=plt.cm.RdBu_r)
ax[1].imshow(im_bf_uneven, cmap=plt.cm.RdBu_r)
ax[0].set_title('original')
ax[1].set_title('uneven illumination')
```

The dark spot at the bottom of the image can be troublesome.  To see this, let's perform Otsu thresholding, which would be our first step in thresholding, to see what happens.

```python
thresh = skimage.filters.threshold_otsu(im_bf)
thresh_uneven = skimage.filters.threshold_otsu(im_bf_uneven)

# Show thresholded images
fig, ax = plt.subplots(1, 2, figsize=(12, 12))
ax[0].imshow(im_bf < thresh, cmap=plt.cm.gray)
ax[1].imshow(im_bf_uneven < thresh_uneven, cmap=plt.cm.gray)
ax[0].set_title('original')
ax[1].set_title('uneven illumination')

# Compare results
print('original: ', thresh)
print('uneven:   ', thresh_uneven)
```

**Background subtraction** is a convenient way to get rid of illumination issues.  To do this, we strongly blur the image so that only the largest length-scale features remain.  We then substract this blurred image from the original image.

We will use `skimage.filters.gaussian_filter` to perform the filtering.  Importantly, `skimage` will automatically convert the original image to a floating point image, assuming it is 16 bit, before filtering.  We therefore need to convert our original image to a floating point image ahead of time so that we can do the appropriate subtraction.

```python
# Convert the uneven image to floating point
im_bf_uneven_float = im_bf_uneven / im_bf_uneven.max()

# Strongly blur the image with a Gaussian filter with stdev of 20 pixels
im_bf_uneven_blur = skimage.filters.gaussian_filter(im_bf_uneven_float, 20)

# Subtract the blurred image from the original
im_bf_new = im_bf_uneven_float - im_bf_uneven_blur

# Display images in false color
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
ax[0,0].imshow(im_bf, cmap=plt.cm.RdBu_r)
ax[0,1].imshow(im_bf_uneven, cmap=plt.cm.RdBu_r)
ax[1,0].imshow(im_bf_uneven_blur, cmap=plt.cm.RdBu_r)
ax[1,1].imshow(im_bf_new, cmap=plt.cm.RdBu_r)
ax[0,0].set_title('original')
ax[0,1].set_title('uneven illumination')
ax[1,0].set_title('strongly blurred')
ax[1,1].set_title('background subtracted')
```

We have now successfully evened out the illumination and we can proceed with segmentation.

##Segmenting the bacteria

As we go about doing the segmentation, it will be instructive to zoom in on a region of the image with some interesting features.  We define the indices of the subimage for use going forward.

```python
# Indices of subimage
r, c = np.indices((200, 300)) + 550

# Look at subimage
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bf[r, c], cmap=plt.cm.gray)
```

###Denoising the image

As we prepare for segementation, we should begin by denoising the image.  We can do this in a variety of ways, but we will do a simple median filter to eliminate possible overexposed pixels, since some cameras are known to have occasional bad pixels.

```python
# Make a cross sequencing element (filter with orthogonal nearest neighbors)
selem = skimage.morphology.disk(1)

# Perform median filter
im_bf_filt = skimage.filters.median(im_bf, selem)

# Look at result
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bf_filt[r, c], cmap=plt.cm.RdBu_r)
```

We may also want to get rid of all the artefacts due to dust on the camera glass.  These are the little circles in the image.  We can use a **total variation filter** for this purpose.  These filters compute local derivatives in the image and try to limit them.  The result is flatter, sometimes "cartoon-like" images.  We will not use this going forward, but only show it here for illustrative purposes.

```python
# Perform a Chambolle total variation filter.
im_bf_tv = skimage.restoration.denoise_tv_chambolle(im_bf_filt, weight=0.005)

# Look at result
fig, ax = plt.subplots(1, 2, figsize=(12, 10))
ax[0].imshow(im_bf_filt[r, c], cmap=plt.cm.RdBu_r)
ax[1].imshow(im_bf_tv[r, c], cmap=plt.cm.RdBu_r)
```

Going forward, we will use the median filtered image.

###Segmenting the bacteria with good ol' thresholding

Assuming we have taken care of any illumination issues that would preclude us from using a global thresholding method, we will go ahead and use a global threshold on the phase images to separate bacteria from background.  Note that this is one of many ways of segmenting the bacteria, and many more sophisticated and power methods are possible.

```python
# Compute Otsu threshold value for median filtered image
thresh_otsu = skimage.filters.threshold_otsu(im_bf_filt)

# Construct thresholded image
im_bw = im_bf_filt < thresh_otsu

# Display images
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
ax[0,0].imshow(im_bf_filt, cmap=plt.cm.gray)
ax[0,1].imshow(im_bw, cmap=plt.cm.gray)
ax[1,0].imshow(im_bf_filt[r, c], cmap=plt.cm.gray)
ax[1,1].imshow(im_bw[r, c], cmap=plt.cm.gray)
```

We see that we have lots of tiny particles that are not considered background.  We will deal with those explicitly in a moment.  Otherwise, we have effectively separated bacteria from background.

Since we will be quantifying fluorescent intensity in individual bacteria, we want to make sure only whole bacteria are considered.  Therefore, we should clear off any bacteria that are touching the border of the image.  There is a very convenient way to do this.

```python
# Clear border with 5 pixel buffer
im_bw = skimage.segmentation.clear_border(im_bw, buffer_size=5)
```

###Quick aside: median filter fluorescent image

Because we will be using the fluorescent image soon, we'll first perform a median filter to make sure we don't have any rogue pixels.

```python
# Perform median filter on fluroscent image
im_fl_filt = skimage.filters.median(im_fl, selem)
```

##Quantifying the segmented regions

We have many white regions in the image, but only some of them are bacteria.  We would like to be able to identify what is bacteria and what is noise.  To this end, we need to compute the number of pixels in each white "island" in the thesholded image.  We can use the `skimage.measure.regionprops` function to do this (and, as we will see, so much more!).  Before we can use it, we have to **label** the image.  That is, we need to assign a unique label to each white region in the thresholded region.

```python
# Label binary image; background kwarg says value in im_bw to be background
im_labeled, n_labels = skimage.measure.label(
                            im_bw, background=0, return_num=True)

# Increment labels by one for future compatibility
im_labeled += 1

# See result (one of the few times it's ok to use rainbow colormap!)
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_labeled, cmap=plt.cm.rainbow)

# Show number of regions
print('Number of individual regions = ', n_labels)
```

We have 351 regions, but only 17 of these are bacteria, as we can see by hand counting.  We want to eliminate the small ones, so we will calculate the area of all of the regions.  `skimage.measure.regionprops` computes the area of each region, in addition to lots of other useful statistics.  Conveniently, through the `intensity_image` keyword argument, we can specify a corresponding intensity image, in this case the fluorescent image.  `skimage.measure.regionprops` then also computes, among other things, average intensity for each of the regions.  Ultimately, we want the integrated intensity, which is just the average intensity times the area.

Note also that we incremented the labels by one.  This is because in `skimage` version `0.11.*`, the background is labeled as `-1`, and the first non-background region is labeled as `0`.  This changes in future versions of `skimage` such that the background is labeled `0`.  `skimage.measure.regionprops` is also expecting background to be labeled zero, with other labels starting at `1`, which is why we incremented the labels by `1`.

```python
# Extract region props
im_props = skimage.measure.regionprops(im_labeled, 
                                       intensity_image=im_fl_filt)
```

That's it!  That's pretty much all there is to it!  We now need to eliminate regions that are too small to be bacteria.  First, let's get an estimate for how big a bacterium is.  We'll look again at our zoomed image.

```python
# Show zoomed in image
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bf_filt[r, c], cmap=plt.cm.gray)
```

Based on the image, I would estimate the bacteria are about 30 pixels long and 10 pixels wide, which is a bit of an underestimate.  So, let's say the cutoff for being a bacterium is about half that, or 150 pixels.  We can use all of the data conveniently stored in `im_props` to clear up our image.

```python
# Make a filtered black and white image
im_bw_filt = im_labeled > 0

# Define cutoff size
cutoff = 150

# Loop through image properties and delete small objects
n_regions = 0
for prop in im_props:
    if prop.area < cutoff:
        im_bw_filt[im_labeled==prop.label] = False
    else:
        n_regions += 1
        
# Look at result
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bw_filt, cmap=plt.cm.gray)

# Show number of regions
print('Number of individual regions = ', n_regions)
```

So, considering only the larger features in the image, we ahve fond the bacteria.

Now, there is still the issue of regions that contain two bacteria.  Again, we look at our zoomed in region.

```python
# Show zoomed in image
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bw_filt[r, c], cmap=plt.cm.gray)
```

The bacteria in the upper right are clearly two different bacteria and may or may not be sisters.  However, the second region from the left is either a bacterium that is dividing, or has just divided.  Either way, you might still want to keep this as a single bacterium.

So, we can test for bacteria that are next to each other, versus those that are in line with each other, as would be the case for a dividing bacterium.  We canuse the eccentricity measure of the region.  According to the [skimage documentation](http://scikit-image.org/docs/dev/api/skimage.measure.html#skimage.measure.regionprops),

>Eccentricity of the ellipse that has the same second-moments as the region. The eccentricity is the ratio of the distance between its minor and major axis length. The value is between 0 and 1.

So, we only want objects with large eccentricity, say above 0.8.

```python
# Loop through image properties and delete small objects
n_regions = 0
for prop in im_props:
    if prop.area < cutoff or prop.eccentricity < 0.8:
        im_bw_filt[im_labeled==prop.label] = False
    else:
        n_regions += 1
        
# Look at result
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(im_bw_filt[r, c], cmap=plt.cm.gray)

# Show number of regions
print('Number of individual regions = ', n_regions)
```

Voila!  Now, we can go ahead and compute the summed intensity of all regions we are interested in.

```python
# Initialize list of intensities of individual bacteria
int_intensity = []

# Loop through regions and compute integrated intensity of bacteria
for prop in im_props:
    if prop.area > cutoff and prop.eccentricity > 0.8:
        int_intensity.append(prop.area * prop.mean_intensity)

# Convert list to NumPy array
int_intensity = np.array(int_intensity)

int_intensity
```

###Overlay of fluorescent image

Finally, we can overlay our fluorescent image with the phase contrast image.  We will take the fluorescent color to be green, and ranging from zero to one.

```python
# Build RGB image by stacking grayscale images
im_rgb = np.dstack(3 * [im_bf_filt / im_bf_filt.max()])

# Make floating point fluorescent image
im_fl_float = im_fl_filt / im_fl_filt.max()

# Substitute the green c
im_rgb[im_bw_filt, 1] = im_fl_float[im_bw_filt]

# Show the result
fig, ax = plt.subplots(1, 2, figsize=(12, 10))
ax[0].imshow(im_rgb)
ax[1].imshow(im_rgb[r, c, :]);
```
